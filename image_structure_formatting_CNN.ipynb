```python
from skimage import io
```

```python
my_image=io.imread('/content/my_image.jpg')
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(my_image)
```

```python
print(my_image.min(),my_image.max())
```

**CONVERT IMAGE DATA TYPE**


```python
from skimage import img_as_float
```

```python
my_img_float=img_as_float(my_image)
plt.imshow(my_img_float)
```

```python
print(my_img_float.min(),my_img_float.max())
```

**MATHEMATICAL OPERATIONS ON IMAGE**


```python
my_dark_image=my_img_float.copy()
my_dark_image=my_dark_image*[.9,.9,.1]
plt.imshow(my_dark_image)
```

**IDEA OF FILTER / KERNEL**

```python
my_filter_image= my_image.copy()
my_filter_image[100:200,400:600,:]=my_filter_image[100:200,400:600,:]*0.5
plt.imshow(my_filter_image)
```
