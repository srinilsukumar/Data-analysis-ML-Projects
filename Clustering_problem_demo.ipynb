Students and their marks for 2 subjects

```python
x = [ 5,  4, 13, 11, 13,  5,  3,  4,  3, 10, 12, 13, 14, 12, 11,  6,  7,  6,  5, 11, 14, 10]
y = [15, 18, 22, 21, 21, 19, 15, 16, 16, 21, 23, 22, 22, 20, 22, 17, 18, 14, 17, 24, 22, 23]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```

```python
data=list(zip(x,y))
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2,n_init="auto")
kmeans.fit(data)
```

```python
c=kmeans.labels_
print(c)
```

```python
plt.scatter(x,y,c=kmeans.labels_)
plt.show()
```

Silhouette Coefficient



```python
from sklearn import metrics
metrics.silhouette_score(data,c,metric='euclidean')
```

Calinski-Harabasz Index



```python
metrics.calinski_harabasz_score(data, c)
```

Davies-Bouldin Index



```python
metrics.davies_bouldin_score(data, c)
```

```python
inertias=[]   ## SSE squeared errors
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,n_init='auto')
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)

print(inertias)
```

```python
plt.plot(range(1,11),inertias,marker='x')
plt.title('Elbow Method')
plt.xlabel('n_clusters')
plt.ylabel('SSE/WCSS')
plt.show()
```

Using Silhouette score

```python
inertias=[]
for i in range(2,11):
  kmeans=KMeans(n_clusters=i,n_init='auto')
  kmeans.fit(data)
  c=kmeans.labels_
  inertia=metrics.silhouette_score(data,c,metric='euclidean')
  inertias.append(inertia)
```

```python
plt.plot(range(2,11),inertias,marker='o')
plt.title('Elbow method')
plt.xlabel('n_clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
